```python
import pandas as pd

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
countries = tables[1]
countries.columns = ['country', '1990', '2005', '2017', '2017%', 'change%', 'land', 'capita', 'total_incl_LUCF', 'total_excl_LUCF']
countries_1 = countries[countries['country'].str.contains("World")==False]
countries_2 = countries_1[countries_1['country'].str.contains("European")==False]

sorted_list = countries_2.sort_values(by=['2017'], ascending = False)
bigger_countries = sorted_list.iloc[:5, [0,1,2,3]]
bigger_countries
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = bigger_countries.columns[1:]
for i, row in bigger_countries.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title("co2 emissions")
plt.xlabel("years")
plt.ylabel("co2 emissions")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.show()

```

```python
import pandas as pd

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
countries = tables[1]
countries.columns = ['country', 'em_1990', '2005', '2017', '2017%', 'change%', 'land', 'capita', 'total_incl_LUCF', 'total_excl_LUCF']
countries_1 = countries[countries['country'].str.contains("World")==False]
countries_2 = countries_1[countries_1['country'].str.contains("European")==False]


countries_2['rel_1990'] = 100
countries_2['rel_2005'] = ((countries_2['2005'] - countries_2['em_1990']) / countries_2['em_1990']) * 100
countries_2['rel_2017'] = ((countries_2['2017'] - countries_2['em_1990']) / countries_2['em_1990']) * 100

countries_2 = countries_2.loc[:209, ['country', 'rel_1990', 'rel_2005', 'rel_2017']]
countries_2_sorted = countries_2.sort_values(by='rel_2017', ascending=False)
countries_2_top_3 = countries_2_sorted.head(3)
countries_2_low_3 = countries_2_sorted.tail(3)

countries_2_top_low = countries_2_top_3.append(countries_2_low_3)




```

```python
import matplotlib.pyplot as plt
import numpy as np

years = countries_2_top_low.columns[1:]
for i, row in countries_2_top_low.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title("co2 emissions relative")
plt.xlabel("years")
plt.ylabel("co2 emissions relative")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.show()

```

```python
import pandas as pd

tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
countries = tables[1]
countries.columns = ['country', 'ch_1990', '2005', '2017', '2017%', 'change%', 'land', 'capita', 'total_incl_LUCF', 'total_excl_LUCF']
countries_1 = countries[countries['country'].str.contains("World")==False]
countries_2 = countries_1[countries_1['country'].str.contains("European")==False]

countries_3 = countries_2.drop(countries_2[countries_2.ch_1990 < 5].index)

countries_3['rel_1990'] = 100
countries_3['rel_2005'] = ((countries_3['2005'] - countries_3['ch_1990']) / countries_3['ch_1990']) * 100
countries_3['rel_2017'] = ((countries_3['2017'] - countries_3['ch_1990']) / countries_3['ch_1990']) * 100

countries_3 = countries_3.loc[:209, ['country', 'rel_1990', 'rel_2005', 'rel_2017']]
countries_3 = countries_3.sort_values(by='rel_2017', ascending=False)
countries_3_top = countries_3.head(3)
countries_3_low = countries_3.tail(3)

countries_3_top_low = countries_3_top.append(countries_3_low)
```

```python
import matplotlib.pyplot as plt
import numpy as np


years = countries_3_top_low.columns[1:]
for i, row in countries_3_top_low.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title("co2 emissions relative 2")
plt.xlabel("years")
plt.ylabel("co2 emissions relative")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.show()
```
